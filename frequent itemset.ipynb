```python
import pandas as pd
import numpy as np
from collections import defaultdict
import matplotlib.pyplot as plt
import os
from datetime import datetime
from operator import itemgetter
import csv
import sys
%matplotlib inline
#plt.style.use('ggplot')
df=pd.read_excel('H:/lottery/ssq.xls',sheet_name='data')
#dfindex=pd.read_csv('C:/Users/xiang.li/Desktop/test1.txt',sep='\s+',low_memory=False)
#df['OPENID']=df['OPENID'].astype('str')
#df=pd.merge(df,dfindex,how='outer',on='pin')
df.columns=['num', 'date', 'r1', 'r2', 'r3', 'r4', 'r5', 'r6', 'b', 'sr1', 'sr2',
       'sr3', 'sr4', 'sr5', 'sr6', 'sales_volume', 'pool', 'first', 'bonus1', 'second', 'bonus2',
       'third', 'bonus3', 'fourth', 'bonus4', 'fifth', 'bonus5', 'sexth', 'bonus6']
#df.tail()
```

```python
X = df.iloc[:,2:8]
each_session_dict = dict((index, frozenset(row)) for index, row in X.iterrows())
single_frequence = defaultdict(int)
frequent_itemsets = {}
min_support = 1
for i in X.columns:
    locals()['dict'+str(i)] = X[i].groupby(X[i]).count().to_dict()
    for k, v in X[i].groupby(X[i]).count().to_dict().items():
        if k in single_frequence.keys():
            single_frequence[k] += v
        else:
            single_frequence[k] = v
frequent_itemsets[1] = dict((frozenset((k,)), v) for (k, v) in single_frequence.items())

```

```python


def find_frequent_itemsets(each_session_dict, k_1_itemsets, min_support):
    counts = defaultdict(int)
    for index, result in each_session_dict.items():
    #reviews is frozenset({...moiveIDs set})
        for itemset in k_1_itemsets:
        #new useage (k,v) in dict.items()------k in dict
        #itemset is the key，namely frozenset({a single movieID}), but itemset will iter with the loop[k_1_itemsets], it will be two movieID ,three ,four...
            if itemset.issubset(result):
            #review is one user's review
            #k_1_itemsets was single movie favoerable over 30, if there are more in one user reveiw, there are  frequency set
                result -= itemset
                for other_reviewed_movie in result:
                #'''for other_reviewed_movie in result - itemset:'''
                #other_reviewed_movie is a single movieID
                    current_superset = itemset | frozenset((other_reviewed_movie,))
                    counts[current_superset] += 1
                    # when itemset and other_reviewed_movie exchanged positon, the count will repeat??? ////////////important???
                    # answer for above: because the reviews is change with itemset
    return dict([(itemset, frequency) for itemset, frequency in counts.items() if frequency > min_support])
```

```python
for k in range(2, 8):
    cur_frequent_itemsets = find_frequent_itemsets(each_session_dict, frequent_itemsets[k-1], min_support)
    frequent_itemsets[k] = cur_frequent_itemsets
    if len(cur_frequent_itemsets) == 0:
        print('no found length {}'.format(k))
        sys.stdout.flush()
        #确保运行时，把缓冲区内容输出到终端，flush解决笔记本文件特定格子的大型循环在代码结束前有时不会输出的问题
        break
    else:
        print('found {} frequent itemsets of length {}'.format(len(cur_frequent_itemsets), k))
        sys.stdout.flush()
#del frequent_itemsets[1]
```
